### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import sklearn.datasets as datasets
from sklearn.metrics import classification_report

# import the functions for visualizing the decision tree
import pydotplus
from IPython.display import Image  
```

### Reading the dataset

```python
# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

### Preparing data for Decision Tree 

```python
y= iris.target
print(y)
```

### Defining the Decision Tree

```python
# Defining the decision tree algorithm

dtree=DecisionTreeClassifier()
dfit = dtree.fit(df,y)

print('\nDecision Tree Classifer Created')

# Acuuracy of decision tree classifier for the above dataset

print('\nAccuracy of DecisionTree classifier : {:.2f}'.format(dtree.score(df, y)))
```

### Visualizing the decision tree

```python
fig = plt.figure(figsize=(25,20))
plt = tree.plot_tree(dfit, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

**Inference:** Hence, a decision tree is created.
